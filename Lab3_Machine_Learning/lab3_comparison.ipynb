```python
import ee
import geemap
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

ee.Authenticate()
ee.Initialize(project='ee-kirschenworks')
```

```python
roi = ee.FeatureCollection('FAO/GAUL/2015/level1') \
    .filter(ee.Filter.eq('ADM0_NAME', 'Thailand')) \
    .filter(ee.Filter.eq('ADM1_NAME', 'Sukhothai'))
```

**training points**

```python
training_points = ee.FeatureCollection('projects/ee-kirschenworks/assets/sukhothai_table_training_points')
```

```python
def mask_s2(image):
    qa = image.select('QA60')
    cloud_bit_mask = 1 << 10
    cirrus_bit_mask = 1 << 11
    mask = qa.bitwiseAnd(cloud_bit_mask).eq(0).And(
        qa.bitwiseAnd(cirrus_bit_mask).eq(0)
    )
    return image.updateMask(mask).divide(10000)

s2 = (ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
      .filterDate('2023-01-01', '2023-12-31')
      .filterBounds(roi)
      .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))
      .map(mask_s2))

image = s2.median().clip(roi)
```

```python
ndvi = image.normalizedDifference(['B8', 'B4']).rename('NDVI')
ndwi = image.normalizedDifference(['B3', 'B8']).rename('NDWI')
ndbi = image.normalizedDifference(['B11', 'B8']).rename('NDBI')
bsi = image.expression(
    '((SWIR + RED) - (NIR + BLUE)) / ((SWIR + RED) + (NIR + BLUE))',
    {
        'SWIR': image.select('B11'),
        'RED': image.select('B4'),
        'NIR': image.select('B8'),
        'BLUE': image.select('B2')
    }
).rename('BSI')

img = image.addBands([ndvi, ndwi, ndbi, bsi])
bands = ['B2', 'B3', 'B4', 'B8', 'B11', 'B12', 'NDVI', 'NDWI', 'NDBI', 'BSI']
```

**sample training data**

```python
samples = img.select(bands).sampleRegions(
    collection=training_points,
    properties=['class'],
    scale=10,
    geometries=True
)

print('Sample count:', samples.size().getInfo())
```

**แบ่ง train / validation**

```python
samples_random = samples.randomColumn('random', 42)

train = samples_random.filter(ee.Filter.lt('random', 0.8))
validation = samples_random.filter(ee.Filter.gte('random', 0.8))

print('Train size:', train.size().getInfo())
print('Validation size:', validation.size().getInfo())
```

**train Random Forest**

```python
rf = ee.Classifier.smileRandomForest(numberOfTrees=100).train(
    features=train,
    classProperty='class',
    inputProperties=bands
)

classified = img.select(bands).classify(rf)
```

**accuracy assessment**

```python
validated = validation.classify(rf)
conf_matrix = validated.errorMatrix('class', 'classification')

print('Confusion Matrix:', conf_matrix.getInfo())
print('Overall Accuracy:', conf_matrix.accuracy().getInfo())
print('Kappa:', conf_matrix.kappa().getInfo())
print('Producer Accuracy:', conf_matrix.producersAccuracy().getInfo())
print('User Accuracy:', conf_matrix.consumersAccuracy().getInfo())
```

**feature importance**

```python
importance = ee.Dictionary(rf.explain().get('importance')).getInfo()

importance_df = pd.DataFrame({
    'Feature': list(importance.keys()),
    'Importance': list(importance.values())
}).sort_values('Importance', ascending=False)

importance_df
```

```python
plt.figure(figsize=(8,5))
plt.bar(importance_df['Feature'], importance_df['Importance'])
plt.xticks(rotation=45)
plt.title('Feature Importance - Random Forest')
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=200)
plt.show()
```

**confusion matrix**

```python
cm = np.array(conf_matrix.getInfo())
classes = ['Urban', 'Water', 'Agriculture', 'Bare_soil']

fig, ax = plt.subplots(figsize=(6,5))
im = ax.imshow(cm)

ax.set_xticks(range(len(classes)))
ax.set_yticks(range(len(classes)))
ax.set_xticklabels(classes, rotation=45, ha='right')
ax.set_yticklabels(classes)

for i in range(len(classes)):
    for j in range(len(classes)):
        ax.text(j, i, cm[i, j], ha='center', va='center')

ax.set_xlabel('Predicted')
ax.set_ylabel('Reference')
ax.set_title('Confusion Matrix')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=200)
plt.show()
```

```python
# Train 3 classifiers
rf = ee.Classifier.smileRandomForest(numberOfTrees=100).train(
    features=train,
    classProperty='class',
    inputProperties=bands
)

cart = ee.Classifier.smileCart().train(
    features=train,
    classProperty='class',
    inputProperties=bands
)

svm = ee.Classifier.libsvm().train(
    features=train,
    classProperty='class',
    inputProperties=bands
)
```

```python
# Evaluate function
def get_metrics(classifier, name):
    validated = validation.classify(classifier)
    matrix = validated.errorMatrix('class', 'classification')
    return {
        'Model': name,
        'Overall Accuracy': matrix.accuracy().getInfo(),
        'Kappa': matrix.kappa().getInfo(),
        'Matrix': matrix
    }

rf_result = get_metrics(rf, 'Random Forest')
cart_result = get_metrics(cart, 'CART')
svm_result = get_metrics(svm, 'SVM')
```

```python
import pandas as pd

comparison_df = pd.DataFrame([
    {'Model': rf_result['Model'], 'Overall Accuracy': rf_result['Overall Accuracy'], 'Kappa': rf_result['Kappa']},
    {'Model': cart_result['Model'], 'Overall Accuracy': cart_result['Overall Accuracy'], 'Kappa': cart_result['Kappa']},
    {'Model': svm_result['Model'], 'Overall Accuracy': svm_result['Overall Accuracy'], 'Kappa': svm_result['Kappa']},
])

comparison_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_cm(matrix, title, filename=None):
    cm = np.array(matrix.getInfo())
    labels = ['Urban', 'Water', 'Agriculture', 'Bare_soil']

    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='viridis',
                xticklabels=labels, yticklabels=labels)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('Reference')
    plt.tight_layout()
    if filename:
        plt.savefig(filename, dpi=200)
    plt.show()
```

```python
plot_cm(rf_result['Matrix'], 'Random Forest Confusion Matrix', 'rf_confusion_matrix.png')
plot_cm(cart_result['Matrix'], 'CART Confusion Matrix', 'cart_confusion_matrix.png')
plot_cm(svm_result['Matrix'], 'SVM Confusion Matrix', 'svm_confusion_matrix.png')
```

```python
best_model = comparison_df.sort_values(by='Overall Accuracy', ascending=False).iloc[0]
print(f"Best model: {best_model['Model']}")
print(f"Overall Accuracy: {best_model['Overall Accuracy']:.3f}")
print(f"Kappa: {best_model['Kappa']:.3f}")
```
